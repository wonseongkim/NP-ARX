```python
import plotly
import ipywidgets
import plotly.express as px
import numpy as np
import pandas as pd
```

```python
#Load the data from the folder specified
# data = pd.read_csv('c:/project/NPARX/oniondata4.csv')
data = pd.read_csv('G:/내 드라이브/academic/KREI PROPHET/KREI 예측자료 20-22/oniondata4.csv')
data.tail() #y is 'bulb weights' and the rest of ups are independent variables
```

```python
data.y.describe()
```

```python
data = data.rename(columns={"date":"ds","y":"y"})
data
```

```python
x_col_lst = ['up1','up2', 'up3', 'up4', 'up5']
y_col_lst = ['y']
X=data[x_col_lst]
Y=data[y_col_lst]

data_pret=pd.concat([data['ds'][2:],X,Y],axis=1)
data_pret=data_pret.reset_index(drop=True)
data_pret
```

```python
#Spilit dataset
cutoff = "2022-01-01"
train = data_pret[data_pret['ds']<cutoff]
test = data_pret[data_pret['ds']>=cutoff]
test
```

```python
col_lst=data_pret.columns
col_lst=col_lst.drop(['ds','y'])
col_lst=list(col_lst)
col_lst
```

```python
#Neural prophet model for bulb weight prediction

import random
import numpy as np
import torch
from neuralprophet import NeuralProphet

seed = 456
random.seed(seed)
np.random.seed(seed)
torch.manual_seed(seed)

# 'AR-Net' in NeuralProphet is a fully connected feed-forward neural network
m = NeuralProphet(
    growth='linear',               # linear growth trend
    yearly_seasonality=True,       # enable yearly seasonality
    weekly_seasonality=True,      # disable weekly seasonality
    daily_seasonality=True,       # disable daily seasonality
    seasonality_mode='multiplicative',   # additive seasonality
    batch_size=64,                 # batch size for training
    epochs=300,                    # number of training epochs
    learning_rate=0.1,             # learning rate for the optimizer
    n_forecasts=1,                 # number of steps to forecast
    n_lags=7,                      # number of historical data points to use
    loss_func='mae',             # Huber loss function
    normalize='soft',            # min-max normalization
    drop_missing=True,           # or,  impute_missing=True
    n_changepoints=100,             # number of changepoints in the trend
    changepoints_range=0.9,        # proportion of the data on which to place changepoints
    #ar_sparsity=0.8,                # sparsity in the AR coefficients
    #num_hidden_layers=4,            # number of hidden layers in AR-Net
    #d_hidden=10,                    # dimension of the hidden layers in AR-Net
)

# Add independent variable(s) and normalize
m = m.add_lagged_regressor(names=col_lst, normalize="soft")

# Train the model
metrics = m.fit(train, freq='D', validation_df=test, progress=None)

# metric checking
print("MAE(Train): ", metrics.MAE.tail(1).item())
print("MAE(Test): ", metrics.MAE_val.tail(1).item())
print("RMSE(Train): ", metrics.RMSE.tail(1).item())
print("RMSE(Test): ", metrics.RMSE_val.tail(1).item())
```

```python
px.line(metrics, y=['MAE', 'MAE_val'], width=800, height=400)
```

```python
forecast = m.predict(test)
fig = m.plot(forecast[['ds', 'y', 'yhat1']])
fig.show()
```
